Q1. What is Min-Max scaling, and how is it used in data preprocessing? Provide an example to illustrate its
application.


```python
"""
Min-Max scaling, also known as normalization, is a data preprocessing technique used to scale and transform 
features within a specific range, typically between 0 and 1. The goal is to ensure that all the features have
the same scale, preventing certain features from dominating the others solely due to their larger magnitude.

The formula for Min-Max scaling is as follows for a single feature:
 
X=(x-xmin)/xmax-xmin


Where:

x is the original value of the feature.

xmin is the minimum value of the feature in the dataset.

xmax is the maximum value of the feature in the dataset.

X is the scaled value of the feature within the range [0, 1].

"""

#An example to illustrate its application Suppose we have a dataset of people's ages:


from sklearn.preprocessing import MinMaxScaler

# Original ages
ages = [25,40,30,20,45,34,19,32,25]

# Create a MinMaxScaler object
scaler = MinMaxScaler()

# Fit and transform the data using the scaler
scaled_ages = scaler.fit_transform([[age] for age in ages])

# Print the scaled ages
for scaled_age in scaled_ages:
    print(scaled_age[0])
```

Q3. What is PCA (Principle Component Analysis), and how is it used in dimensionality reduction? Provide an
example to illustrate its application.


```python
"""
Principal Component Analysis (PCA) is a technique used for reducing the dimensionality of data while retaining its 
important information. It works by finding new directions in the data that capture the most variability. Here's a
brief summary of the steps and its application:

1-Center Data: Move data points so they're centered around the origin.

2-Calculate Covariance Matrix: Compute the matrix that shows how features vary together.

3-Find Eigenvectors and Eigenvalues: Identify the directions of maximum variance and the amount of variance along those directions.

3-Select Principal Components: Choose the eigenvectors with the highest eigenvalues as the new coordinate axes.

4-Project Data: Transform data by projecting it onto the selected principal components.
"""


from sklearn.decomposition import PCA
import numpy as np

# Original data
data = np.array([[2.5, 2.4],
                 [0.5, 0.7],
                 [2.2, 2.9],
                 [1.9, 2.2],
                 [3.1, 3.0],
                 [2.3, 2.7],
                 [2.0, 1.6],
                 [1.0, 1.1],
                 [1.5, 1.6],
                 [1.1, 0.9]])

# Create a PCA object with 1 principal component
pca = PCA(n_components=1)

# Fit the data and transform it to the lower-dimensional space
lower_dim_data = pca.fit_transform(data)

# Print the lower-dimensional data
print(lower_dim_data)
```

Q8. For a dataset containing the following features: [height, weight, age, gender, blood pressure], perform
Feature Extraction using PCA. How many principal components would you choose to retain, and why?

```python
"""
The decision of how many principal components to retain in PCA (Principal Component Analysis) depends on 
the specific characteristics of your dataset and the goals you have in mind. Here are the steps you can follow
to determine the number of principal components to retain:

1-Standardization: Start by standardizing your dataset. PCA is sensitive to the scale of the features, so it's
important to ensure that all features have the same scale.

2-Compute Covariance Matrix: Calculate the covariance matrix of the standardized features. The covariance matrix
represents the relationships between features.

3-Compute Eigenvalues and Eigenvectors: Calculate the eigenvalues and eigenvectors of the covariance matrix. These 
eigenvectors are the principal components, and their corresponding eigenvalues indicate the amount of variance explained 
by each component.

4-Explained Variance: Calculate the explained variance ratio for each principal component. This is the proportion of the total
variance explained by each component. It helps you understand how much information each component retains.

5-Scree Plot: Plot the explained variance ratios against the principal components. The scree plot can help you visualize the
"elbow point," which indicates where the explained variance starts to level off.

6-Cumulative Explained Variance: Calculate the cumulative explained variance by summing up the explained variance ratios. This
will help you decide how many principal components to retain to capture a desired amount of variance.

7-Threshold: Set a threshold for the amount of variance you want to retain. Common thresholds are 95% or 99% of the total variance.
Choose the number of principal components that collectively explain at least that amount of variance.
"""


#import Libraries
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Sample data (replace this with your actual dataset)
data = np.array([[170, 65, 30, 1, 120],
                 [160, 55, 25, 0, 130],
                 [180, 70, 35, 1, 125],
                 [175, 67, 40, 0, 110]
                ])

#Convert data into dataframe
df=pd.DataFrame(data,columns=['height','weight','age','gender','blood pressure'])


# Step 1: Standardize the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)

# Step 2: Calculate covariance matrix
covariance_matrix = np.cov(scaled_data, rowvar=False)

# Step 3: Compute eigenvalues and eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(covariance_matrix)

# Step 4: Calculate explained variance ratio
explained_variance_ratio = eigenvalues / np.sum(eigenvalues)

# Step 5: Scree plot
plt.plot(range(1, len(explained_variance_ratio) + 1), explained_variance_ratio, marker='o')
plt.xlabel('Principal Component')
plt.ylabel('Explained Variance Ratio')
plt.title('Scree Plot')
plt.show()

# Step 6: Calculate cumulative explained variance
cumulative_explained_variance = np.cumsum(explained_variance_ratio)

# Step 7: Set a threshold and determine the number of principal components to retain
threshold = 0.95  # You can adjust this threshold
num_components_to_retain = np.argmax(cumulative_explained_variance >= threshold) + 1

print("Number of principal components to retain:", num_components_to_retain)

# Perform PCA with the chosen number of components
pca = PCA(n_components=num_components_to_retain)
principal_components = pca.fit_transform(scaled_data)

# Now principal_components contains the extracted features
principal_components
```
